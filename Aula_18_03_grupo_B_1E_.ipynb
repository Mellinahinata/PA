```python
# Operadores aritméticos

# Fazer um programa que o usuário digita 02 números
# Saída mostrar operações aritméticas

# Entrada
a=int(input('Digite um número'))
b=int(input('Digite um número'))

# Processamento
soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a /b
potencia = a ** b
resto = a % b

# Saída
print("")
print("------------------------ OPERADORES ARITMÉTICOS -----------------------")
print("")
print("-------------------------------SOMA -----------------------------------")
print(f"o resultado da SOMA de (a) MAIS (b) é igual a : {soma} " )
print("-----------------------------------------------------------------------")

print("")
print("------------------------------- SUBTRACAO -----------------------------")
print(f"o resultado da SUBTRACAO de (a) MENOS (b) é igual a : {subtracao} " )
print("-----------------------------------------------------------------------")


print("")
print("------------------------------- MULTIPLICACAO -------------------------")
print(f"o resultado da MULTIPLICACAO de (a) MULTIPLICADO (b) é igual a : {multiplicacao} " )
print("-----------------------------------------------------------------------")


print("")
print("------------------------------- DIVISAO -------------------------------")
print(f"o resultado da DIVISAO de (a) DIVIDIDO (b) é igual a : {divisao} " )
print("-----------------------------------------------------------------------")


print("")
print("------------------------------- POTENCIA -----------------------------------")
print(f"o resultado da POTENCIA de (a) ELEVADO (b) é igual a : {potencia} " )
print("-----------------------------------------------------------------------")



```

```python
# Fazer um programa que converte km em metros

# Entrada
km = int(input("digite a quantidade de km  "))

# Processamento
km = km * 1000

# Saída
print(f"a quantidade em metros é igual a {km}")
```
