```python
# 1 - fazer um programa que identifique se o número é par ou ímpar
# entrada
numero = int(input("digite um número:  "))

if numero % 2 == 0:
    print(f"o número {numero} é par")
else:
   print(f"o número {numero} é impar")
```

```python
# 2- um programa que solicite dois números e mostre qual é o maior (ou se são iguais).
# entrada
numero1 = int(input("digite um número: "))
numero2 = int(input("digite outro número:  "))

if numero1 > numero2:
    print(f"o número {numero1} é maior que o número {numero2}")

elif numero1<numero2:
    print(f"o número {numero2} é maior que o número {numero1}")

else:
    print(f" o número {numero1} é igual ao número {numero2}")

```

```python
# 3 - Escrever um programa que Peça a nota de um aluno e informe:
# Aprovado (nota ≥ 7)
# Recuperação (5 ≤ nota < 7)
# Reprovado (nota < 5)


# Entrada
nota1 = float (input("digite sua nota do primeiro bimestre:  "))
nota2 = float (input("digite sua nota do segundo bimestre:  "))
nota3 = float (input("digite sua nota do terceiro bimestre:  "))
nota4 = float (input("digite sua nota do quarto bimestre:  "))

# Processamento

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:
  print (f"a média das suas notas foi {media}, você foi aprovado.")

elif media >= 5:
  print (f"a média das suas notas foi {media}, você terá que fazer recuperação.")

else:
  print (f"a média da suas notas foi {media}, você foi reprovado.")

```

```python
# 4- Receba a idade e classifique: Criança (0–12), Adolescente (13–17), Adulto (18–59) e Idoso (60+)
# entrada
idade = int (input("digite sua idade:  "))

# processamento
if idade <= 12:
  print (f"você é uma criança")

elif idade <= 17:
  print (f"você é um adolescente")

elif idade <= 59:
  print (f"você é um adulto")

else:
    print (f"você é um idoso")


```

```python
# Peça o valor de uma compra: ≥ 100: 10% de desconto, ≥ 50 e < 100: 5% de desconto , < 50: sem desconto
# entrada
valor = float (input("digite o valor da compra:  "))

# processamento
if valor >= 100:
  valor_final = valor * 0.90
  print(f"O valor final da compra com 10% de desconto é: R$ {valor_final:.2f}")

elif valor >= 50:
  valor_final = valor * 0.95
  print(f"O valor final da compra com 5% de desconto é: R$ {valor_final:.2f}")

else:
  valor_final = valor
  print(f"O valor final da compra é: R$ {valor_final:.2f}, não possui desconto")
```
