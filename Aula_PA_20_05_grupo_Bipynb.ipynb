```python
# Escrever um programa que recebe 04 notas do aluno, calcula a média e imprimi aprovado, recuperação ou reprovado.

# regras:

# utilizar while
# nota maior ou igual a 7 aprovado
# nota maior ou igual a 5 recuperação
# nota menor que 5 reprovado

notas = 0
bimestre = 1
soma = 0

try:

  while bimestre <= 4:
    notas = float (input(f"digite sua nota do {bimestre}º bimestre: "))
    soma = soma + notas / 4
    bimestre +=1

  if soma >= 7:
    print(f"aprovado! sua média foi {soma} ")

  elif soma >= 5:
    print (f"recuperação! sua média foi {soma}")

  else:
    print (f"reprovado! sua média foi {soma}")
except:
  print ("digite apenas números")
```

```python
# você foi contratado para desenvolver um sistema simples de caixa eletrônico.

# o sistema deverá funcionar através de um menu interativo utilizando a estrutura de repetição do while.

# o programa deve iniciar com um saldo inicial de R$ 1.000,00 e permitir que o usuário escolhas diferentes operações até decidir sair do sistema

# funcionalidades:

# 1 - ver saldo;
# 2 - depositar dinheiro
# 3 - sacar dinheiro
# 4 - sair

# regras:

# ver saldo
# ao escolher essa opção, o sistema deverá mostrar o saldo atual da conta;

# depositar dinheiro
# o usuáro deverá informar um valor para o depósito.
# o valor digitado deve ser somado ao saldo

# sacar dinheiro
# o usuário deverá informar o valor do saque.
# se houver o suficiente - realizar o saque.
# se não houver saldo suficiente - mostrar mensagem de "saldo insuficiente".

# sair

# o programa deverá encerrar apenas quando o usuário escolher a opção.

saldo = 1000.00
opcao = 0

while opcao != 4:

  print("\n--- CAIXA ELETRÔNICO ---")
  print("opção - 1 - Saldo Atual")
  print("opção - 2 - Depositar Dinheiro")
  print("opção - 3 - Sacar Dinheiro")
  print("opção - 4 - Sair")
  try:
    opcao = int (input("digite a opção desejada: "))
  except:
    print("Entrada inválida. Por favor, digite um número.")
    continue

  if opcao == 1:
    print (f"Seu saldo atual é de R$ {saldo}")

  elif opcao == 2:
    try:
      deposito = float (input("Digite o valor do depósito: "))
      if deposito > 0:
        saldo += deposito
        print (f"Depósito de R$ {deposito} realizado com sucesso. Seu novo saldo é de R$ {saldo}")
      else:
        print("O valor do depósito deve ser positivo.")
    except:
      print("Entrada inválida para o depósito. Por favor, digite um número.")

  elif opcao == 3:
    try:
      saque = float (input("Digite o valor do saque: "))
      if saque > 0:
        if saque <= saldo:
          saldo -= saque
          print (f"Saque de R$ {saque} realizado com sucesso. Seu novo saldo é de R$ {saldo}")
        else:
          print ("Saldo insuficiente.")
      else:
        print("O valor do saque deve ser positivo.")
    except:
      print("Entrada inválida para o saque. Por favor, digite um número.")

  elif opcao == 4:
    print ("Saindo!")
  else:
    print ("Opção inválida, por favor digite uma opção entre 1 e 4.")
```
