```python
# jogo menu inteligente # criar um sistema que funcione até o usuário escolher sair.

# menu
# 1 - par/impar
# 2 - positivo/negativo
# 3 - maior
# 4 - sair

opcao = 0
while opcao != 4:

  print("--- JOGO MENU INTELIGENTE ---")
  print("opção - 1 - par/impar")
  print("opção - 2 - positivo/negativo")
  print("opção - 3 - maior")
  print("opção - 4 - Sair")

  opcao = int(input("digite a opção desejada: "))
  if opcao == 1:
      numero = int(input("digite um número: "))
      if numero % 2 == 0:
        print(f"o número {numero} é par")
      else:
        print(f"o número {numero} é ímpar")

  elif opcao == 2:
      numero = int(input("digite um número: "))
      if numero > 0:
        print(f"o número {numero} é positivo")
      elif numero < 0:
        print(f"o número {numero} é negativo")
      else:
        print(f"o número {numero} é neutro (zero)")
  elif opcao == 3:
      numero1 = int(input("Digite o primeiro número: "))
      numero2 = int(input("Digite o segundo número: "))
      if numero1 > numero2:
        print(f"O maior número é {numero1}")
      elif numero2 > numero1:
        print(f"O maior número é {numero2}")
      else:
        print("Os números são iguais.")

  elif opcao == 4:
    print("Saindo! obrigado por jogar!")

  else:
    print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")
```

```python
# urna eletrônica

# criar um programa de votação com 3 candidatos

# votar candidato A
# votar candidato B
# votar candidato C
# votar branco
# votar nulo
# encerrar votação

# o sistema deve repetir até encerrar a votação
# contar votos de cada candidato
# contar votos brancos e nulos
# ao encerra, mostrar:

# total de votos
# total de votos por candidato
# total votos nulos
# vencedor da eleição

# regras extras

# se houver empate entre candidatos : mostrar eleição empatada
# se não houver votos válidos, mostrar : "nenhum candidato recebeu votos"

candidato_1 = 0
candidato_2 = 0
candidato_3 = 0
brancos = 0
nulos = 0
total_votos = 0

while True:
    print("--- URNA ELETRÔNICA ---")
    print("[ 1 ] Candidato palhaço caçarola")
    print("[ 2 ] Candidato tirulipa")
    print("[ 3 ] Candidato gordox ")
    print("[ 4 ] Voto em Branco")
    print("[ 5 ] Voto Nulo")
    print("[ 01 ] Encerrar Votação")


    voto = input("Digite o número correspondente ao seu voto: ")

    if voto == "1":
        candidato_1 += 1
        total_votos += 1
        print("Voto para Candidato 1 confirmado! ")
    elif voto == "2":
        candidato_2 += 1
        total_votos += 1
        print("Voto para Candidato 2 confirmado! ")
    elif voto == "3":
        candidato_3 += 1
        total_votos += 1
        print("Voto para Candidato 3 confirmado! ")
    elif voto == "4":
        brancos += 1
        total_votos += 1
        print("Voto em BRANCO confirmado! ")
    elif voto == "5":
        nulos += 1
        total_votos += 1
        print("Voto NULO confirmado! ")
    elif voto == "0":

        confirmar = input("Tem certeza que deseja encerrar a votação? (S/N): ")

        if confirmar == "S":
            print("\nVotação encerrada com sucesso! Processando resultados...")
            break
    else:
        print(" Opção inválida! Digite um número de 1 a 5, ou 0 para encerrar.")

print("\n" + "="*40)
print("---RESULTADO FINAL---")
print("="*40)
print(f"Total de votos computados: {total_votos}")
print(f"• Votos Brancos: {brancos}")
print(f"• Votos Nulos: {nulos}")
print("-" * 40)
print(f"• Candidato 1: {candidato_1} voto(s)")
print(f"• Candidato 2: {candidato_2} voto(s)")
print(f"• Candidato 3: {candidato_3} voto(s)")
print("-" * 40)
votos_validos = candidato_1 + candidato_2 + candidato_3
if votos_validos == 0:
    print(" Nenhum candidato recebeu votos válidos.")
else:
    maior_voto = max(candidato_1, candidato_2, candidato_3)
    vencedores = []
    if candidato_1 == maior_voto:
        vencedores.append("Candidato 1")
    if candidato_2 == maior_voto:
        vencedores.append("Candidato 2")
    if candidato_3 == maior_voto:
        vencedores.append("Candidato 3")

    #
    if len(vencedores) > 1:
        print(f" Houve um empate entre os candidatos: {', '.join(vencedores)}!")
    else:
        print(f" O vencedor da eleição é: {vencedores[0]}!")
```
