```python
# jogo de adivinhação

# o usuário terá 03 chances para adivinhar um número de 0 a 10.

# se o número secreto for maior ou menor que o palpite o programa irá imprimir uma mensagem informando.

# informar se o palpite está quente ou frio.

import random

numero_secreto = random.randint(0, 10)
tentativas = 0
max_tentativas = 3

print("--- JOGO DE ADIVINHAÇÃO ---")
print(f"digite um número entre 0 e 10. Você possui {max_tentativas} tentativas.\n")

ganhou = False

while tentativas < max_tentativas:
    try:
        tentativa = int(input("digite um número de 0 a 10: "))
        if not (0 <= tentativa <= 10):
            print("digite apenas um número entre 0 e 10.")
            continue
    except ValueError:
        print("número inválido, digite um número inteiro.")
        continue

    tentativas += 1

    if tentativa == numero_secreto:
        print(f"você acertou o número secreto {numero_secreto} em {tentativas} tentativa(s), parabéns! 😲")
        ganhou = True
        break
    else:
        diferenca = abs(tentativa - numero_secreto)
        if diferenca <= 2:
            print("está muito quente!")
        elif diferenca <= 4:
            print("está quente!")
        else:
            print("está frio!")

        if tentativa < numero_secreto:
            print("O número secreto é maior.")
        else:
            print("O número secreto é menor.")

        print(f"Você ainda tem {max_tentativas - tentativas} tentativa(s).\n")

if not ganhou:
    print(f"suas tentativas acabaram! O número secreto era {numero_secreto}.")
```
