```python
# QUESTÃO 02 - soma dos números

# Solicite ao usuário um número inteiro positivo.

# Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.

numero = int(input("digite um número inteiro positivo:  "))
soma = 0

for i in range(1, numero + 1):
  soma += i

  print(f"a soma dos números de 1 até {numero} é: {soma}")
```

```python
# QUESTÃO 03 – Tabuada

# Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.
numero = int(input("digite um número inteiro:  "))

for i in range(1,11):
  print(f"{numero} x {soma} = {numero * i}")
```

```python
# QUESTÃO 04 – Fatorial

# Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.

# Observação:O programa deve validar se o número informado é positivo.

numero = int(input("digite um número inteiro positivo:  "))
if numero > 0:
  print ("o número é positivo, pode prosseguir")
else:
  print ("digite apenas números positivos")
fatorial = 1

for i in range(1, numero + 1):
  fatorial *= i

  print(f"o fatorial de {numero} é: {fatorial}")
```

```python
# QUESTÃO 05 – Caixa Eletrônico
# Um caixa eletrônico possui notas de: R$ 100, R$ 50, R$ 20, R$ 10, R$ 5, R$ 2 e R$ 1

# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.

print("------CAIXA ELETRÔNICO-------")

valor = int(input("Digite o valor do saque: "))

if valor >= 100:
    notas_100 = 0
    while valor >= 100:
        notas_100 += 1
        valor -= 100
    print(f"{notas_100} nota(s) de R$100")

if valor >= 50:
    notas_50 = 0
    while valor >= 50:
        notas_50 += 1
        valor -= 50
    print(f"{notas_50} nota(s) de R$50")

if valor >= 20:
    notas_20 = 0
    while valor >= 20:
        notas_20 += 1
        valor -= 20
    print(f"{notas_20} nota(s) de R$20")

# R$ 10 notes
if valor >= 10:
    notas_10 = 0
    while valor >= 10:
        notas_10 += 1
        valor -= 10
    print(f"{notas_10} nota(s) de R$10")

if valor >= 5:
    notas_5 = 0
    while valor >= 5:
        notas_5 += 1
        valor -= 5
    print(f"{notas_5} nota(s) de R$5")

if valor >= 2:
    notas_2 = 0
    while valor >= 2:
       notas_2 += 1
       valor -= 2
    print(f"{notas_2} nota(s) de R$2")

if valor >= 1:
    notas_1 = 0
    while valor >= 1:
        notas_1 += 1
        valor -= 1
    print(f"{notas_1} nota(s) de R$1")
```
