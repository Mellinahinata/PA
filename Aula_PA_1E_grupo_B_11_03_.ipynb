# Escrever um programa que faz conversão de temperatura

Entrada: temperatura em celsius

Processamento: converter para Fahrenheit

Saída: resultado.

Formula: F= (C x 9/5) + 32

```python
# Calculo de conversão

# Entrada: temperatura em celsius
celsius = int (input("digite um numero "))

# Processamento
fahrenheit = (celsius * 9/5) + 32

# Saída
print(f"o valor em fahrenheit é igual a:{fahrenheit}")

```
