```python
# cofre digital

# um cofre possui um código composto por 5 números inteiros.

# desenvolva um programa que:

# leia os 5 números utilizando um laço de repetição.

# ao final informar:

# 1 - quantos números são pares.
# 2 - quantos números são ímpares
# 3 - o maior número.
# 4 - a soma de todos os números.
# 5 - a média dos números
# 6 - o menor número.
# 7 - quantos números são maiores que a média.

numero = []
for i in range(5):
    while True:
        try:
            num = int(input(f"digite o {i+1}º número inteiro: "))
            numero.append(num)
            break
        except ValueError:
            print("por favor, digite um número inteiro.")


if numero:
    pares = sum(1 for num in numero if num % 2 == 0)
    impares = len(numero) - pares

    maior = max(numero)

    soma = sum(numero)

    media = soma / len(numero)

    menor = min(numero)

    maior_media = sum(1 for num in numero if num > media)

    print("- Resultados -")
    print(f"Números digitados: {numero}")
    print(f"Quantidade de números pares: {pares}")
    print(f"Quantidade de números ímpares: {impares}")
    print(f"Maior número: {maior}")
    print(f"Soma dos números: {soma}")
    print(f"Média dos números: {media:.2f}")
    print(f"Menor número: {menor}")
    print(f"Quantidade de números maiores que a média: {maior_media}")
else:
    print("Nenhum número foi digitado.")
```
