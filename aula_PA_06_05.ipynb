```python
# 1- contagem simples, o usuário digita um número e o programa imprimi do zero até o número.
número = int(input("Digite um número: "))
for i in range (número + 1):
  print (i)
```

```python
# 2- contagem regressiva, o usuário digita um número e o programa imprimi do número até o zero.
numero= int(input("digite um número:  "))
for i in range(numero, -1, -1):
  print (i)
```

```python
# 3- números pares, o usuário digita um número e o programa imprimi os números pares.
numero= int(input("digite um número:  "))
for i in range(2, numero, 2):
  print (i)
```

```python
# 4- números ímpares, o usuário digita um número e o programa imprimi os números ímpares.
numero= int(input("digite um número:  "))
for i in range(1, numero,2):
  print (i)
```

```python
# 5- soma, o usuário digita um número e o programa soma os números do zero até o número.
numero= int(input("digite um número:  "))
soma = 0
for i in range(numero, -1, -1):
  soma += i
print("A soma dos números do zero até", numero, "é:", soma)
```

```python
# 6- tabuada, o usuário digita um número e o programa imprimi a tabuada do número.
numero= int(input("digite um número:  "))
for i in range(1, 11):
  print(f"{numero} x {i} = {numero * i}")
```

```python
# 7- contar letras, o usuário digita uma palavra e o programa conta as letras da palavra.
palavra = input("Digite uma palavra: ")
contador = 0
for letra in palavra:
  contador += 1
print("A palavra tem", contador, "letras.")
```

```python
# 8- contar vogais, o usuário digita uma palavra e o programa conta as vogais.
palavra = input("digite uma palavra:  ")

vogais= 0

for letra in palavra:
  if letra in "aeiouAEIOU":
    vogais += 1
    print (vogais)
```
