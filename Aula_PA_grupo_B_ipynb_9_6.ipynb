```python
# faça um programa que calcule e acumule a soma dos números digitados pelo usuário
continuar = 's'
total_acumulado = 0

while continuar.lower() == 's':
  numero1 = int(input("digite o primeiro número:  "))
  numero2 = int(input("digite o segundo número:  "))
  soma = 0

  if numero1 <= numero2:
    for i in range(numero1, numero2 + 1):
      soma += i
  else:
    for i in range(numero2, numero1 + 1):
      soma += i

  print(f"A soma dos números entre {numero1} e {numero2} é: {soma}")
  total_acumulado += soma
  print(f"A soma total acumulada é: {total_acumulado}")

  continuar = input("Deseja continuar? (s/n): ")

print("Programa encerrado. A soma final acumulada foi:", total_acumulado)
```

```python
# solicite ao usuário 4 notas e calcule média
notas = 0
bimestre = 1
soma = 0

try:

  while bimestre <= 4:
    notas = float (input(f"digite sua nota do {bimestre}º bimestre: "))
    soma = soma + notas / 4
    bimestre +=1

  if soma >= 7:
    print(f"aprovado! sua média foi {soma} ")

  elif soma >= 5:
    print (f"recuperação! sua média foi {soma}")

  else:
    print (f"reprovado! sua média foi {soma}")
except:
  print ("digite apenas números")

```

```python
# faça um programa que recebe valores e quantidade de produtos e no final exiba o valor total da compra
valor_total = 0
produtos = 0
valor_produto = 0

while produtos <= 0:
  produtos = int(input("digite a quantidade de produtos: "))

  if produtos > 0:
    for i in range(produtos):
      valor_produto = float(input(f"digite o valor do produto {i+1}: "))
      valor_total += valor_produto
    print(f"O valor total da compra é de {valor_total:.2f} R$")
```
