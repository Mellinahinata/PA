```python
# Escrever um programa que digita o nome e as 04 notas do bimestre, calcula média e mostra
# aprovado, recuperação ou reprovado
# nota igual ou maior que 7 aprovado
# nota maior ou igual 5 e 7 recuperação;
# nota menor que 5 reprovado;

# Entrada

nome = str (input("digite seu nome:  "))
nota1 = float (input("digite sua nota do primeiro bimestre:  "))
nota2 = float (input("digite sua nota do segundo bimestre:  "))
nota3 = float (input("digite sua nota do terceiro bimestre:  "))
nota4 = float (input("digite sua nota do quarto bimestre:  "))

# Processamento

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:
  print (f"{nome}, a média das suas notas foi {media}, você foi aprovado.")

elif media >= 5:
  print (f"{nome}, a média das suas notas foi {media}, você terá que fazer recuperação.")

else:
  print (f"{nome}, a média da suas notas foi {media}, você foi reprovado.")

```

```python
# escrever um programa que recebe os valores de 03 lados e veriica se é um triângulo válido e classifique o tipo de triângulo;

# Verificar se o triâmgulo é válido;
# * a soma de dois lados quaisquer deve ser sempre maior que o terceiro lado

# Classificar tipos de triângulos;

# Equilátero: Três lados e três ângulos internos iguais.
# Isósceles: Dois lados iguais e os ângulos da base também iguais.
# Escaleno: Três lados e três ângulos com medidas diferentes.

# Entrada
lado1 = int (input("digite o primeiro lado:  "))
lado2 = int (input("digite o segundo lado:  "))
lado3 = int (input("digite o terceiro lado:  "))

# Processamento
if (lado1 + lado2) > lado3 and (lado1 + lado3) > lado2 and (lado2 + lado3) > lado1:
  if lado1 == lado2 == lado3:
    print (f"Seu triângulo é equilátero.")
  elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
    print (f"Seu triângulo é isósceles.")
  else:
    print (f"Seu triângulo é escaleno.")
else:
  print (f"Os lados digitados não formam um triângulo válido.")
```
